```python
#P3
data1 = pd.Series([10, 20, 30, 40, 50])
data2 = pd.Series([5, 4, 3, 2, 1])

#Operasi aritmatika antar series
jumlah = data1.add(data2)        
selisih = data1.subtract(data2) 
produk = data1.multiply(data2) 
bagi = data1.divide(data2)      

print("Hasil Penjumlahan:\n", jumlah)
print("Hasil Pengurangan:\n", selisih)
print("Hasil Perkalian:\n", produk)
print("Hasil Pembagian:\n", bagi)
```
